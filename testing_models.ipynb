```python
import os
import cv2
import copy
import time
import tqdm
import warnings
import matplotlib
import numpy as np
import pandas as pd
from PIL import Image
import seaborn as sns
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn import model_selection, metrics, preprocessing
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score
```

## Reading the test set

```python
path_folder = 'C:\\Users\\HP\\Downloads\\computervition_project\\archive (2)\\TEST'

size = 124
data_test = []
label_test = []
for folder in tqdm.tqdm(os.listdir(path_folder)):
      for files in tqdm.tqdm(os.listdir(os.path.join(path_folder, folder))):
            if files.endswith('jpg'):
                label_test.append(folder)
                img_path = os.path.join(path_folder, folder, files)
                img = cv2.imread(img_path, cv2.IMREAD_COLOR)
                im = cv2.resize(img, (size, size))
                data_test.append(im)
```

## Normalizing data

```python
data_test = np.array(data_test)
data_test = data_test.astype('float32') / 255
```

## Encode the label

```python
label_encoder = preprocessing.LabelEncoder()
label_test = label_encoder.fit_transform(label_test)
```

## one hot encode

```python
from keras.utils import to_categorical
label_test_one_hot = to_categorical(label_test, num_classes=3)
```

## Testing the model with VGG19 using 10 epochs

```python
model = tf.keras.models.load_model('my_model19.h5')
```

```python
predictions = model.predict(data_test)
```

```python
# Convert one-hot encoded labels to class indices
true_labels = np.argmax(label_test_one_hot, axis=1)

# Convert predicted probabilities to class indices
predicted_labels = np.argmax(predictions, axis=1)

# Calculate overall accuracy
acc = accuracy_score(true_labels, predicted_labels)
print("The accuracy is:", acc)

print()
# Calculate ROC-AUC for each class
class_indices = [0, 1, 2]
for index in class_indices:
    roc_auc = roc_auc_score(label_test_one_hot[:, index], predictions[:, index])
    print("ROC-AUC Score for Class {}: {:.4f}".format(index, roc_auc))

print()
roc_auc = roc_auc_score(label_test_one_hot, predictions, multi_class='ovr')
print("ROC-AUC Score: ", roc_auc)
```
